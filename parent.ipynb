```python
import pandas as pd
import numpy as np
import random
from datetime import datetime, timedelta

# Seed for reproducibility
np.random.seed(42)
random.seed(42)

# Generate common data
dates = [datetime(2026, 1, random.randint(1, 31)).strftime(random.choice(['%Y-%m-%d', '%d/%m/%Y', '%m-%d-%Y'])) for _ in range(200)]
products = ['T-Shirt', 'Jeans', 'Dress', 'Jacket', 'Shoes', 'Hat', 'Scarf', 'Belt']
categories = ['Clothing', 'Apparel', 'Footwear', 'Accessories', 'clothing', 'apparel ']  # Messy free-text
locations = ['London', 'Manchester', 'Birmingham', 'Edinburgh', 'london store', None]
quantities = np.random.randint(1, 10, 200)
prices = np.round(np.random.uniform(5.0, 100.0, 200), 2)
revenues = quantities * prices

# POS (in-store) - inconsistent columns
pos_df = pd.DataFrame({
    'Sale_Date': dates[:100],
    'ProductName': random.choices(products, k=100),
    'Category': random.choices(categories, k=100),
    'Store_Location': random.choices(locations, k=100),
    'Qty': quantities[:100],
    'Price': prices[:100],
    'Total_Revenue': revenues[:100]
})
# Add mess: duplicates, missing
pos_df = pd.concat([pos_df, pos_df.sample(10)])  # Duplicates
pos_df.iloc[::10, pos_df.columns.get_loc('Qty')] = np.nan  # Missing values

# Online (Shopify) - different columns
online_df = pd.DataFrame({
    'date': dates[100:],
    'item': random.choices(products, k=100),
    'prod_category': random.choices(categories, k=100),
    'location': ['Online'] * 100,  # All online
    'quantity_sold': quantities[100:],
    'unit_price': prices[100:],
    'revenue': revenues[100:],
    'channel': ['shopify'] * 100
})
# Add mess: duplicates, missing, extra noise
online_df = pd.concat([online_df, online_df.sample(5)])
online_df.iloc[::15, online_df.columns.get_loc('quantity_sold')] = np.nan
online_df['prod_category'] = online_df['prod_category'].apply(lambda x: x.strip() + random.choice(['', ' typo']))

# Save to CSVs
pos_df.to_csv('pos_sales_messy.csv', index=False)
online_df.to_csv('online_sales_messy.csv', index=False)

print("Generated: pos_sales_messy.csv and online_sales_messy.csv")
```
